## 문제 1:  다음 이진 검색 트리에 대해 답하시오.

#1) 이 이진 검색 트리의 리프 노드는 총 몇개인가?
-> [3개] 키가 7, 17, 70인 노드들이므로 3개이다.

#2) 현재 높이 4를 유지하면서 키를 추가한다면 가능한 리프 노드의 최대 개수는 몇개인가?
-> [8개] 이진 트리에서 높이가 h일 때, 리프 노드의 최대 개수는 2^(h-1)이다. 높이 4를 유지한다고 하였으므로 2^(4-1)=8이다. 따라서 존재할 수 있는 리프 노드의 최대 개수는 8개이다.

#3) 높이 4인 이진 검색 트리가 포화 이진 트리 모양이 되면 리프 노드와 리프 노드를 제외한 나머지 노드(내부 노드라 한다)의 수는 몇 개나 차이가 나는가?
-> [1개] 높이를 h라 할 때, 전체 노드 수는 2^h - 1 이고 리프 노드 수는 2^(h-1)이다. 내부 노드의 수 = 전체 노드 수 - 리프 노드 수이다. h가 4인 것을 대입하면 내부 노드 수 = 15 - 8 = 7이다. 따라서 리프 노드 수와 내부 노드 수의 차이는 1개이다.

#4) 위 3)의 성질은 높이가 ℎ인 모든 포화 이진 트리에 대해 성립하는가?
-> [성립한다.] 내부 노드가 1개 늘어날 때마다 리프 노드도 1개씩 늘어나서 모든 포화 이진 트리에서 리프 노드의 수 = 내부 노드 수 + 1 이 성립하기 때문이다.

## 문제 2:  임의의 완전 이진 트리에 대한 다음 물음에 답하시오.

#1) 트리의 총 노드 수가 𝑛이면 리프 노드의 수는 몇 개인가?
-> 	[⌈𝑛/2⌉ 개이다.] 완전 이진 트리는 마지막 레벨을 제외하고는 꽉 차 있고, 왼쪽부터 노드들이 채워지는 구조이기 때문에 자식이 있는 노드와 없는 노드가 절반으로 나뉘게 된다.

#2) 위 1)에서 리프 노드의 수와 내부 노드의 수는 어떤 관계인가?
-> 내부 노드는 리프 노드가 아닌 노드이므로 뺄셈으로 계산할 수 있는 관계이다. 또한 리프 노드는 내부 노드보다 1이 많거나 같다.

## 문제 9:  𝑛개의 원소로 구성되는 이진 검색 트리의 가능한 최대 높이와 최소 높이는 어떻게 되는가?

-> [최대 높이: n, 최소 높이: log (n+1)] 최대 높이일 때는 트리가 한쪽으로만 계속 늘어진 모양이기 때문에 노드의 개수만큼인 n이 높이가 된다. 최소 높이일 때는 트리가 완전히 균형 잡힌 경우이기 때문에 로그 수준이 된다.
